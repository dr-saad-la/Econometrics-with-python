## Accessing Federal Reserve Economic Data (FRED)

**FRED (Federal Reserve Economic Data)** is one of the most comprehensive sources for US economic data, providing access to over 800,000 economic time series.

### Example Usage

In this section we will import the US GDP from the `FRED` source using `DataReader()` from pandas-datareader package

````python
# first: import the data module as web (if you haven't done that yet)
import pandas_datareader.data as web


# Define date range 
start_date = datetime.datetime(2018, 1, 1)
end_date = datetime.datetime(2023, 1, 1)

# Fetch US GDP (quarterly data)
gdp = web.DataReader('GDP', 'fred', start_date, end_date)

# Display the first few rows
print("US GDP Data:")
print(gdp.head())
````

````python
# Check the data info
print()
print(gdp.info())
````

### Fetching US Unemployement Rate

You can fetch the US unemployment monthly rate using `DataReader()` in the same way you did above:

````python
# Fetch US Unemployment Rate (monthly data)
unemployment = web.DataReader('UNRATE', 'fred', start_date, end_date)
````

````python
# Display the first few rows
print("\nUS Unemployment Rate:")
print(unemployment.head())
````

### Consumer Price Index for All Urban Consumers

The Consumer Price Index for All Urban Consumers: All Items (CPIAUCSL) is a price index of a basket of goods and services paid by urban consumers. Percent changes in the price index measure the inflation rate between any two time periods. [source](https://fred.stlouisfed.org/series/CPIAUCSL)

````python
# Fetch Consumer Price Index (monthly data)
cpi = web.DataReader('CPIAUCSL',
                     'fred',
                     start_date,
                     end_date)

# Display the first few rows
print("\nConsumer Price Index:")
print(cpi.head())
````

### Consumer Price Indices for China

The Consumer Price Indices (CPIs, HICPs), COICOP 1999: Consumer Price Index: Total for China [CHNCPIALLMINMEI]. [Source](https://fred.stlouisfed.org/series/CHNCPIALLMINMEI)

````python
# Fetch Consumer Price Index (monthly data)
china_cpi = web.DataReader('CHNCPIALLMINMEI',
                     'fred',
                     start_date,
                     end_date)
````

````python
# Display the first few rows
print("\nConsumer Price Index:")
print(china_cpi.head())
````

### Economic Indicators Visualization

The first step after aquiring the data is to explore it visually, so let's visualize these economic indicators:

````python
# Create a figure with 3 subplots
fig, axes = plt.subplots(1, 3, figsize=(12, 6))

# Plot GDP
gdp.plot(ax=axes[0], title='US Gross Domestic Product (GDP)')
axes[0].set_ylabel('Billions of Dollars')
axes[0].grid(True)

# Plot Unemployment Rate
unemployment.plot(ax=axes[1], title='US Unemployment Rate')
axes[1].set_ylabel('Percent')
axes[1].grid(True)

# Plot CPI
cpi.plot(ax=axes[2], title='Consumer Price Index for All Urban Consumers')
axes[2].set_ylabel('Index 1982-1984=100')
axes[2].grid(True)

plt.tight_layout()
plt.show()
````

## Accessing Multiple Series at Once

We can fetch multiple related series from FRED in one operation:

````python
# Get multiple interest rate series
interest_rates = web.DataReader(['DFF', 'DGS10', 'DGS30'],
                                'fred',
                                start_date,
                                end_date)

# Rename columns for clarity
interest_rates.columns = ['Federal Funds Rate', '10-Year Treasury', '30-Year Treasury']

# Display the data
print("Interest Rates:")
print(interest_rates.head())
````

````python
# Plot the data
interest_rates.plot(title='US Interest Rates', figsize = (10, 6))
plt.ylabel('Percent')
plt.grid(True)
plt.show()
````

## World Bank Development Indicators

The World Bank provides extensive economic, social, and development data for countries around the world.

Note that when fetching one or more economic indicators for more than on country you will get a panel data:

````python
# import the world bank module 
import pandas_datareader.wb as wb

# Fetch GDP per capita for selected countries
gdp_per_capita = pdr.wb.download(indicator='NY.GDP.PCAP.CD', 
                                country=['US', 'CA', 'GB', 'DE', 'JP', 'CN'],
                                start=2014,
                                 end=2024)

# Check the first few observations
gdp_per_capita.head(15)
````

````python
# Check the last few observations
gdp_per_capita.tail(15)
````

````python
# Check the available indexes
print(gdp_per_capita.index.values)
````

````python
# Reshape the data for easier plotting
gdp_pc_unstack = gdp_per_capita.reset_index().pivot(index='year',
                                                    columns='country',
                                                    values='NY.GDP.PCAP.CD')

# Display the data
print("GDP per Capita:")
print(gdp_pc_unstack.head())
````

````python
# Plot the data
gdp_pc_unstack.plot(title='GDP per Capita', figsize = (12, 6))
plt.ylabel('Current US$')
plt.grid(True)
plt.show()
````

## Best Practices for Using pandas-datareader

When working with pandas-datareader, consider these best practices:

1. **Error Handling**: Always implement proper error handling to manage API rate limits and connection issues:

```python
try:
    data = web.DataReader('SERIES_ID', 'SOURCE', start_date, end_date)
except Exception as e:
    print(f"Error fetching data: {e}")
    # Handle the error appropriately
```

2. **Data Validation**: Always check the data after retrieval:

```python
def validate_data(df, series_name):
    print(f"Validating {series_name}:")
    print(f"- Shape: {df.shape}")
    print(f"- Date range: {df.index.min()} to {df.index.max()}")
    print(f"- Missing values: {df.isna().sum().sum()}")
    
    if df.isna().any().any():
        print("- Warning: Contains missing values")
    
    return df
```

3. **Caching Data**: For frequently used series, consider caching the data:

```python
import os

def get_data_with_cache(series_id, source, start, end, cache_dir='data_cache'):
    """Fetch data with caching to reduce API calls"""
    if not os.path.exists(cache_dir):
        os.makedirs(cache_dir)
    
    cache_file = os.path.join(cache_dir, f"{source}_{series_id}.pkl")
    
    if os.path.exists(cache_file):
        df = pd.read_pickle(cache_file)
        if df.index.min() <= start and df.index.max() >= end:
            return df[(df.index >= start) & (df.index <= end)]
    
    # If cache doesn't exist or doesn't cover the requested date range, fetch from API
    df = web.DataReader(series_id, source, start, end)
    df.to_pickle(cache_file)
    return df
```

## API Deprecation and Alternatives

Some data sources in pandas-datareader may be deprecated or have limited functionality. For more robust access, consider these alternatives:

1. **FRED data**: Use the `fredapi` package
2. **World Bank data**: Use `wbdata` package
3. **Yahoo Finance data**: Use `yfinance` package
4. **Quandl data**: Use the `quandl` package

Example of using alternatives:

```python
# Using fredapi instead of pandas-datareader
from fredapi import Fred
fred = Fred(api_key='YOUR_API_KEY')  # API key is optional
gdp_data = fred.get_series('GDP')

# Using yfinance instead of pandas-datareader
import yfinance as yf
apple_data = yf.download('AAPL', start='2020-01-01', end='2023-01-01')
```
