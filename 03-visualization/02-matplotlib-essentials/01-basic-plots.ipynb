## 1. Basic Line Plots

Line plots are essential for visualizing time series data in economics. Let's create some example time series data and learn different ways to plot it.

```python
# Set the seed for reproducibility
np.random.seed(2)
# Generate sample time series data
dates = pd.date_range(start='2020-01-01',
                      end='2023-12-31',
                      freq='ME')

values = np.cumsum(np.random.randn(len(dates))) + 100          # Random walk with drift

# Create basic line plot
plt.figure(figsize=(10, 6))
plt.plot(dates, values, 'b-', label='Stock Price')

plt.title('Basic Line Plot: Stock Price Over Time')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(visible = True)
plt.xticks(rotation=30)

plt.show()
```

### Line Plot Styles and Colors

Matplotlib offers various line styles and colors. Here's how to use them:

```python
# Generate multiple series
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x) * np.exp(-x/5)

plt.figure(figsize=(12, 6))
plt.plot(x, y1, 'b-', label='Solid Blue Line')
plt.plot(x, y2, 'r--', label='Dashed Red Line')
plt.plot(x, y3, 'g:', label='Dotted Green Line')

plt.title('Different Line Styles and Colors', 
         font="georgia",
         fontweight="bold",
         fontsize=14,
         color = "brown")

plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
```

## 2. Scatter Plots

Scatter plots are crucial for visualizing relationships between variables in economics.

```python
# Generate correlated data
np.random.seed(123)
n_points = 100
x = np.random.normal(0, 2, n_points)
y = 0.5 * x + np.random.normal(0, 0.5, n_points)

plt.figure(figsize=(10, 6))
plt.scatter(x, y, alpha=0.5, c='blue')

plt.title('Scatter Plot: Relationship Between Variables',
         font="georgia",
         fontweight="bold",
         fontsize=14,
         color = "brown")
plt.xlabel('Independent Variable')
plt.ylabel('Dependent Variable')
plt.grid(True)
plt.show()
```

### Scatter Plot with Size and Color Variation

```python
# Generate data with additional dimension
np.random.seed(123)

n_points = 50
x = np.random.normal(0, 2, n_points)
y = 0.5 * x + np.random.normal(0, 0.5, n_points)
sizes = np.random.uniform(20, 200, n_points)
colors = np.random.uniform(0, 1, n_points)


plt.figure(figsize=(10, 6))
scatter = plt.scatter(x, y, 
                      s=sizes,
                      c=colors,
                      alpha=0.6,
                      cmap='viridis')


plt.colorbar(scatter)
plt.title('Scatter Plot with Varying Size and Color',
         font="georgia",
         fontweight="bold",
         fontsize=14,
         color = "brown")

plt.xlabel('Independent Variable')
plt.ylabel('Dependent Variable')
plt.grid(True)

plt.show()
```

## 3. Bar Charts

Bar charts are useful for comparing categorical data or time-discrete measurements.

```python
# Create sample data
categories = ['Q1', 'Q2', 'Q3', 'Q4']
values = np.random.uniform(50, 100, 4)

plt.figure(figsize=(10, 6))
plt.bar(categories, values, color='skyblue')
plt.title('Quarterly Revenue')
plt.xlabel('Quarter')
plt.ylabel('Revenue (millions)')

# Add value labels on top of each bar
for i, v in enumerate(values):
    plt.text(i, v + 1, f'{v:.1f}', ha='center')

plt.show()
```

### Grouped Bar Charts

```python
# Data for multiple categories
years = ['2020', '2021', '2022', '2023']
q1_data = np.random.uniform(70, 100, 4)
q2_data = np.random.uniform(70, 100, 4)

x = np.arange(len(years))
width = 0.35

plt.figure(figsize=(12, 6))
plt.bar(x - width/2, q1_data, width, label='Region A', color='skyblue')
plt.bar(x + width/2, q2_data, width, label='Region B', color='lightcoral')

plt.title('Regional Comparison by Year')
plt.xlabel('Year')
plt.ylabel('Value')
plt.xticks(x, years)
plt.legend()
plt.show()
```

## 4. Multiple Plots and Subplots

Often in economic analysis, we need to display multiple related plots together.

```python
# Create figure with 2x2 subplots
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(12, 10))

# Plot 1: Line plot
x = np.linspace(0, 10, 100)
ax1.plot(x, np.sin(x))
ax1.set_title('Line Plot')
ax1.grid(True)

# Plot 2: Scatter plot
ax2.scatter(np.random.normal(0, 1, 50), np.random.normal(0, 1, 50), alpha=0.5)
ax2.set_title('Scatter Plot')
ax2.grid(True)

# Plot 3: Bar plot
ax3.bar(['A', 'B', 'C', 'D'], np.random.uniform(1, 10, 4))
ax3.set_title('Bar Plot')

# Plot 4: Histogram
ax4.hist(np.random.normal(0, 1, 1000), bins=30, alpha=0.7)
ax4.set_title('Histogram')

plt.tight_layout()
plt.show()
```

## 5. Plot Customization

Let's explore various ways to customize plots for professional presentations and publications.

```python
# Create a professional-looking plot with multiple customizations
plt.figure(figsize=(12, 7))

# Generate data
x = np.linspace(0, 10, 100)
y1 = np.sin(x) * np.exp(-x/10)
y2 = np.cos(x) * np.exp(-x/10)

# Create plots with custom styles
plt.plot(x, y1, 'b-', linewidth=2, label='Series 1')
plt.plot(x, y2, 'r--', linewidth=2, label='Series 2')

# Customize the plot
plt.title('Professional Plot Example', fontsize=14, pad=15)
plt.xlabel('Time', fontsize=12)
plt.ylabel('Value', fontsize=12)

# Customize grid
plt.grid(True, linestyle='--', alpha=0.7)

# Customize legend
plt.legend(loc='upper right', frameon=True, framealpha=0.9)

# Customize spines
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Add text annotation
plt.annotate('Peak', xy=(1.4, 0.86), xytext=(2.5, 1.02),
             arrowprops=dict(facecolor='black',
                             edgecolor='black',
                             shrink=0.05,
                             width=1.5,
                            headlength=17,
                            headwidth=10))

plt.tight_layout()
plt.show()
```
